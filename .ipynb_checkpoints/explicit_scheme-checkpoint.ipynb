# Excplicit differential equaton solving

```python
D=0.86                                      # D coefcient  
L=20                                        # x interval (0,20)
T=100                                       # t interval (0,100)
```

```python
dx=1
dt=0.05
print(dt,dx**2/2*D)
```

```python
dx=0.5
dt=0.025
print(dt,dx**2/2*D)
```

```python
dx=0.25
dt=0.0125
print(dt,dx**2/2*D)
```

```python
import math
import pylab
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
Nx=int(L/dx)+1                                         
Nt=int(T/dt)+1
print(dt,dx**2/2*D)

startBorder_0_x = lambda t,x: x*(60-x)/30   # Starting border u(t=0,x)
startBorder_t_0 = lambda t,x: 30            # Left border line u(t,x=0)
endBorder_t_L = lambda t,x: 0               # Right border line u(t,x=L)

pi=math.pi
ck=lambda k: 60/((k*pi)**3)*(1-(k*pi)**2*(1+math.cos(k*pi)))
analytic=lambda t,x,D,L: (0-30)*x/L+30+sum([ck(k)*math.exp((-((k*pi)**2)*D*t/L**2))*math.sin(k*pi*x/L) for k in range(1,10)])

u=[]
w=[]

u.append([startBorder_0_x(0,x*dx) for x in range(Nx)]) #init starting border
w.append([startBorder_0_x(0,x*dx) for x in range(Nx)])
t=0
x=0
a=D*dt/(dx**2)

for n in range (Nt-1):
    u.append([])
    w.append([])
    t+=dt
    temp=startBorder_t_0(t+dt,0)
    u[n+1].append(temp)
    w[n+1].append(temp)
    for j in range(1,Nx-1):
        x+=dx
        u[n+1].append(u[n][j] + a*(u[n][j+1] - 2*u[n][j] + u[n][j-1]))
        w[n+1].append(analytic(t,x,D,L))
    x=0
    temp=endBorder_t_L(t+dt,1)
    u[n+1].append(temp)  
    w[n+1].append(temp)
    
ts=[t*dt for t in range(Nt)]
xs=[x*dx for x in range(Nx)]    

xs,ts = np.meshgrid(xs,ts)
```

```python
fig = pylab.figure()
ax = Axes3D(fig)
u=np.array(u)
w=np.array(w)
ax.plot_wireframe(ts, xs, u)
ax.view_init(50,200)
ax.set_xlabel('T')
ax.set_ylabel('X')
ax.set_zlabel('F(T,X)')
pylab.show()
```

```python
fig = pylab.figure()
ax = Axes3D(fig)
u=np.array(u)
w=np.array(w)
ax.plot_wireframe(ts, xs, w)
ax.view_init(50,200)
ax.set_xlabel('T')
ax.set_ylabel('X')
ax.set_zlabel('F(T,X)')
pylab.title("3D plot analytical solution")
pylab.show()
```

```python
norm1=[]
ts1=ts[1:]
for i in range(1,Nt):
    norm1.append((sum([(u[i][j]-w[i][j])**2 for j in range(Nx)])/Nx)**(1/2))
```

```python
norm2=[]
ts2=ts[1:]
for i in range(1,Nt):
    norm2.append((sum([(u[i][j]-w[i][j])**2 for j in range(Nx)])/Nx)**(1/2))
```

```python
norm3=[]
ts3=ts[1:]
for i in range(1,Nt):
    norm3.append((sum([(u[i][j]-w[i][j])**2 for j in range(Nx)])/Nx)**(1/2))
```

```python

fig = pylab.figure()
pylab.subplot(321)
pylab.title("RMSE dx=1,dt=0.05")
pylab.plot(ts1,norm1)

pylab.subplot(322)
pylab.title("RMSE dx=0.5,dt=0.025")
pylab.plot(ts2,norm2)

pylab.subplot(313)
pylab.plot(ts3,norm3)
pylab.title("RMSE dx=0.25,dt=0.0125")
pylab.show()
```
